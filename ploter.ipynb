```python
from IXIdataset import IXIDataset
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
```

```python
dataset = IXIDataset('/data_hdd/users/lisikuang/IXI/T1/valid/')
loader = DataLoader(dataset, batch_size=4, shuffle=True, drop_last=True, num_workers=16)
```

```python
imgs, depth = next(iter(loader))
imgs = imgs[0].squeeze()
depth = depth[0].squeeze()
print(imgs.shape)
print(depth.shape)

fig, axs = plt.subplots(1, 3, figsize=(10, 3))
axs[0].imshow(imgs[0], cmap="gray")
axs[0].set_title('lower_image')
axs[0].axis('off')
axs[1].imshow(imgs[1], cmap="gray")
axs[1].set_title('upper_image')
axs[1].axis('off')
axs[2].imshow(imgs[2], cmap="gray")
axs[2].set_title('label_image')
axs[2].axis('off')

plt.show()
```
